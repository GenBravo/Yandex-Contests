Формат ввода

Ключами словаря, записанного в файле <input.json>, являются номера вводимых строк, а значениями — следующие типы преобразований:

    10 – из фраз, записанных в данной строке, необходимо выбрать только те, у которых не менее двух различных гласных;
    20 – выбрать фразы с четной длиной;

```python
import json

words = [0]
vowels = 'aeuioy'

# line = input()
filename = input()

# break_reading_data = False
# while (not break_reading_data):
#     line = input()
#     if line[0] == '#':
#         break_reading_data = True
#         break
#     words.append(line.split('_'))

with open('yandex contest/' + filename) as json_file:
# with open(filename) as json_file:
    data = json.load(json_file)

num_of_lines = len(data)

for i in range(num_of_lines):
    line = input()
    words.append(line.split('_'))

output_dict = {}

for key in data.keys():
    output_dict[key] = []
    
    if data[key] == '10':
        for word in words[int(key)]:
            dif_vowel_count = 0
            for i in vowels:
                if i in word:
                    dif_vowel_count += 1
            if dif_vowel_count > 1:
                if output_dict[key] != []:
                    output_dict[key].append(word)
                else:
                    output_dict[key] = [word]
                    
    elif data[key] == '20':
        for word in words[int(key)]:
            is_even = (len(word) % 2 == 0)
            if is_even:
                if output_dict[key] != []:
                    output_dict[key].append(word)
                else:
                    output_dict[key] = [word]
                    
    elif data[key] == '30':
        output_dict[key] = words[int(key)]

    output_dict[key].sort(reverse=True)
 
# print(data)
# print(output_dict)

with open("output.json", "w") as outfile: 
    json.dump(output_dict, outfile, indent=4, sort_keys=True)
```
