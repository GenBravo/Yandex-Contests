### A. Покраска деревьев

Ограничение времени 	1 секунда
Ограничение памяти 	64Mb

Ввод 	стандартный ввод или input.txt
Вывод 	стандартный вывод или output.txt

Вася и Маша участвуют в субботнике и красят стволы деревьев в белый цвет. Деревья растут вдоль улицы через равные промежутки в 1 метр. Одно из деревьев обозначено числом ноль, деревья по одну сторону занумерованы положительными числами 1,2 и т.д., а в другую — отрицательными −1,−2 и т.д.

Ведро с краской для Васи установили возле дерева P, а для Маши — возле дерева Q. Ведра с краской очень тяжелые и Вася с Машей не могут их переставить, поэтому они окунают кисть в ведро и уже с этой кистью идут красить дерево. Краска на кисти из ведра Васи засыхает, когда он удаляется от ведра более чем на V метров, а из ведра Маши — на M метров. Определите, сколько деревьев может быть покрашено.

**Формат ввода**
В первой строке содержится два целых числа P и V — номер дерева, у которого стоит ведро Васи и на сколько деревьев он может от него удаляться.
В второй строке содержится два целых числа Q и M — аналогичные данные для Маши.
Все числа целые и по модулю не превосходят 108.

**Формат вывода**
Выведите одно число — количество деревьев, которые могут быть покрашены.


**Пример**
Ввод
Вывод

0 7
12 5

	

25


````python
P, V = map(int, input().split())
Q, M = map(int, input().split())

vasya_low = P - V
vasya_top = P + V
masha_low = Q - M
masha_top = Q + M
inter_start = None
inter_end = None

lowest_of_two = 'vasya'
highest_of_two = 'vasya'

intersection_present = False
intersection_len = 0

if masha_low < vasya_low:
    lowest_of_two = 'masha'
    
if masha_top > vasya_top:
    highest_of_two = 'masha'
   

if lowest_of_two == 'vasya':
    if masha_low <= vasya_top:
        intersection_present = True
        inter_start = masha_low       
elif lowest_of_two == 'masha':
    if vasya_low <= masha_top:
        intersection_present = True
        inter_start = vasya_low

        
if (lowest_of_two == highest_of_two) and intersection_present:
    if lowest_of_two == 'vasya':
        intersection_len = 2 * M + 1
    elif lowest_of_two == 'masha':
        intersection_len = 2 * V + 1
        
elif (lowest_of_two != highest_of_two) and intersection_present:
    if lowest_of_two == 'vasya':
        intersection_len = vasya_top - inter_start + 1
    elif lowest_of_two == 'masha':
        intersection_len = masha_top - inter_start + 1

# print(f'vasya_low {vasya_low} \
# vasya_top {vasya_top} \
# masha_low {masha_low} \
# masha_top {masha_top} \
# inter_start {inter_start} \
# inter_end {inter_end} \
# intersection_present {intersection_present} \
# intersection_len {intersection_len} \
# lowest_of_two {lowest_of_two} \
# highest_of_two {highest_of_two}')
        
final_sum = (2 * V + 1) + (2 * M + 1) - intersection_len

print (final_sum)
````

### C. Форматирование файла

Ограничение времени 	1 секунда
Ограничение памяти 	256Mb

Ввод 	стандартный ввод или input.txt
Вывод 	стандартный вывод или output.txt

Петя - начинающий программист. Сегодня он написал код из n строк.

К сожалению оказалось, что этот код трудно читать. Петя решил исправить это, добавив в различные места пробелы. А точнее, для i-й строки ему нужно добавить ровно ai пробелов.

Для добавления пробелов Петя выделяет строку и нажимает на одну из трёх клавиш: Space, Tab, и Backspace. При нажатии на Space в строку добавляется один пробел. При нажатии на Tab в строку добавляются четыре пробела. При нажатии на Backspace в строке удаляется один пробел.

Ему хочется узнать, какое наименьшее количество клавиш придётся нажать, чтобы добавить необходимое количество пробелов в каждую строку. Помогите ему!

**Формат ввода**
Первая строка входных данных содержит одно целое положительное число n(1≤n≤105) – количество строк в файле.
Каждая из следующих n строк содержит одно целое неотрицательное число ai(0≤ai≤109) – количество пробелов, которые нужно добавить в i-ю строку файла.

**Формат вывода**
Выведите одно число – минимальное количество нажатий, чтобы добавить в каждой строке необходимое количество пробелов.

**Пример**
Ввод
Вывод

5
1
4
12
9
0

8

Примечания
В примере можно:

    1 раз нажать на Space в первой строке.
    1 раз нажать на Tab на второй строке.
    3 раза нажать на Tab в третьей строке.
    2 раза нажать на Tab и один раз нажать на Space в четвёртой строке.
    Ничего не нажимать в пятой строке.

В итоге получается 1+1+3+3=8 нажатий. Можно доказать, что нельзя добавить необходимое количество пробелов за 7 нажатий или меньше.


````python
num_of_lines = int(input())

lines=[]
presses_needed = 0

for i in range(num_of_lines):
    line = int(input())
    lines.append(line)
    
def find_num_of_presses(n):
    if n == 0:
        return 0
    elif n < 3:
        return n
    elif n == 3:
        return 2
    elif n == 4:
        return 1
    else:
        return (n//4 + find_num_of_presses(n%4))
    
    
for i in range(num_of_lines):
    presses_needed += find_num_of_presses(lines[i])
    
print(presses_needed)
````

На шахматной доске стоят слоны и ладьи, необходимо посчитать, сколько клеток не бьется ни одной из фигур.

Шахматная доска имеет размеры 8 на 8. Ладья бьет все клетки горизонтали и вертикали, проходящих через клетку, где она стоит, до первой встретившейся фигуры. Слон бьет все клетки обеих диагоналей, проходящих через клетку, где он стоит, до первой встретившейся фигуры.

````python
chess_board = []

# chess_board = ['********',
# '********',
# '*R******',
# '********',
# '********',
# '********',
# '********',
# '********']

# chess_board = ['********',
# '********',
# '******B*',
# '********',
# '********',
# '********',
# '********',
# '********']

# chess_board = ['********',
# '*R******',
# '********',
# '*****B**',
# '********',
# '********',
# '********',
# '********'
# ]

# chess_board = ['********',
# '***B****',
# '**BRB***',
# '***B****',
# '********',
# '********',
# '********',
# '********']

squares_beat = set()
squares_remaining = 64
squares_that_contain_pieces = set()

for i in range(8):
    chess_board.append(input().strip())

for row in range(8):
    for col in range(8):
        if chess_board[row][col] != '*':
            squares_that_contain_pieces.add((row,col))

    
def rook_beats_squares(row, col):
    set_of_squares = set()
    
    #up
    for i in range(row-1,-1,-1):
        if (i,col) not in squares_that_contain_pieces:
            set_of_squares.add((i,col))
        else:
            break
    #down
    for i in range(row+1,8,1):
        if (i,col) not in squares_that_contain_pieces:
            set_of_squares.add((i,col))
        else:
            break
    #right
    for i in range(col+1,8,1):
        if (row,i) not in squares_that_contain_pieces:
            set_of_squares.add((row,i))
        else:
            break
    #left
    for i in range(col-1,-1,-1):
        if (row,i) not in squares_that_contain_pieces:
            set_of_squares.add((row,i))
        else:
            break
            
    return set_of_squares


def bishop_beats_squares(row, col):
    set_of_squares = set()
    
    #up-right
    row_ = row - 1
    col_ = col + 1
    while (row_ > -1) and (col_ < 8):
        if (row_, col_) not in squares_that_contain_pieces:
            set_of_squares.add((row_, col_))
            row_ -= 1
            col_ += 1
        else:
            break
        
    #up-left
    row_ = row - 1
    col_ = col - 1
    while (row_ > -1) and (col_ > -1):
        if (row_, col_) not in squares_that_contain_pieces:
            set_of_squares.add((row_, col_))
            row_ -= 1
            col_ -= 1
        else:
            break
            
    #down-right
    row_ = row + 1
    col_ = col + 1
    while (row_ < 8) and (col_ < 8):
        if (row_, col_) not in squares_that_contain_pieces:
            set_of_squares.add((row_, col_))
            row_ += 1
            col_ += 1
        else:
            break    
    
    #down-left
    row_ = row + 1
    col_ = col - 1
    while (row_ < 8) and (col_ > -1):
        if (row_, col_) not in squares_that_contain_pieces:
            set_of_squares.add((row_, col_))
            row_ += 1
            col_ -= 1
        else:
            break    

    return set_of_squares


# for row in range(8):
#     for col in range(8):
#         if chess_board[row][col] == 'R':
#             squares_beat = squares_beat.union(rook_beats_squares(row, col))
#         elif chess_board[row][col] == 'B':
#             squares_beat = squares_beat.union(bishop_beats_squares(row, col))

for row, col in squares_that_contain_pieces:
        if chess_board[row][col] == 'R':
            squares_beat = squares_beat.union(rook_beats_squares(row, col))
        elif chess_board[row][col] == 'B':
            squares_beat = squares_beat.union(bishop_beats_squares(row, col))


squares_beat = squares_beat.union(squares_that_contain_pieces)           

squares_remaining -= len(squares_beat)
    
print(squares_remaining)
````
